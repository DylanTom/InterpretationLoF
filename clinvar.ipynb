```python
import pandas as pd
```

```python
df3 = pd.read_csv('maybeFinal.csv')
df3 = pd.DataFrame(df3)
```

```python
df3.head()
```

```python
df4 = pd.DataFrame(columns = ['CHROM','POS','ID','REF','ALT', 'QUAL','FILTER','INFO'])
```

```python
columns = ['CHROM','POS','REF','ALT']
data = []
for x, y in df3.iterrows():
    values = []
    for i, j in df1.iterrows():
        if y[3] == j[8]:
            values.append(j[0])
            values.append(j[1])
            values.append(j[3])
            values.append(j[4])
    dictionary = dict(zip(columns, values))
    data.append(dictionary)
```

```python
df4 = df4.append(data, True)
```

```python
header = """##fileformat=VCFv4.1
##fileDate=2020-08-30
##source=ClinVar
##reference=GRCh38
#CHROM POS ID REF ALT QUAL FILTER INFO
"""
output_VCF = "myfile2.vcf"
with open(output_VCF, 'w') as vcf:
    vcf.write(header)

df4.to_csv(output_VCF, sep="\t", mode='a', index=False)
```
