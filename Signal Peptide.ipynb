```python
from pprint import pprint

fasta_file = "signalPeptide.txt"

def parse_fasta(fname):
    contents = []
    with open(fname, "r") as fh:
        for line in fh:
            contents.append(line)
    return contents

fasta = parse_fasta(fasta_file)
pprint(fasta)
```

```python
type(fasta)
```

```python
fasta1 = [ x for x in fasta if ">" not in x ]
fasta2 = [i.strip() for i in fasta1]
```

```python
fasta_final = []
for x in fasta2:
    if x not in fasta_final:
        fasta_final.append(x)
```
